```python
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)
```

```python
import pandas as pd
# read data and put it in a dataframe
# 在 google 工作表載入 gsheets
gsheets = gc.open_by_url('https://docs.google.com/spreadsheets/d/1-LBvMluvrbeiM2Zq4mKAldT80jf-o8fEuOmqarLH2xM/edit?usp=sharing')
```

```python
# 從 gsheets 的 All-whiteboard-device 載入 sheets
sheets = gsheets.worksheet('記帳本').get_all_values()
# 將 sheets1 資料載入 pd 的 DataFrame 進行分析
df = pd.DataFrame(sheets[1:], columns=sheets[0])
# 取得最前面的5筆資料
df.head()
```

```python
import datetime
```

```python
# 讓使用者輸入資料
date_str = input("請輸入日期 (格式：YYYY-MM-DD): ")
# 檢查日期格式是否正確
datetime.datetime.strptime(date_str, '%Y-%m-%d')

time_str = input("請輸入時間 (格式：HH:MM): ")
# 檢查時間格式是否正確
datetime.datetime.strptime(time_str, '%H:%M')

item = input("請輸入品項: ")
amount = float(input("請輸入金額: "))
```

```python
# 創建一個新的 DataFrame 來代表你要新增的資料
new_data = pd.DataFrame([{
    '日期': date_str,
    '時間': time_str,
    '分類':'遊戲',
    '品項': item,
    '錢': amount,
    '誰':'我',
    '哪':'家',
    '方式':'信用卡',
    '備註':'-',

}])

# 使用 concat() 將新資料合併到舊的 df 中
df = pd.concat([df, new_data], ignore_index=True)
```

```python
# 將 DataFrame 中的所有 NaN 值替換為 0
df = df.fillna(0)

# 顯示 DataFrame 的前幾行以確認更改
df.head()
```

```python
df
```

```python
# 將 DataFrame 轉換成列表的列表 (list of lists)，這是 gspread 支援的資料格式
data_to_write = df.values.tolist()

# 第一步：取得工作表物件
worksheet = gsheets.worksheet('記帳本')

# 第二步：只將 data_to_write 的最後一列資料寫入工作表
# 注意：data_to_write[-1] 是一個列表，append_rows 預期接收一個列表的列表，所以需要用 [] 包裹起來
worksheet.append_rows(values=[data_to_write[-1]], value_input_option='USER_ENTERED')
print("最新的一筆資料已成功寫入 Google 記帳本")
```
